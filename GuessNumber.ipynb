```python
import random
numb = random.randint(1,100)
guess_numb = int(input("Enter the number"))
while True:
  if numb == guess_numb:
    print("YOU ARE RIGHT! GREAT JOB")
    break
  elif guess_numb > numb:
    print("YOUR GUESS IS HIGH, TRY AGAIN")
    guess_numb = int(input("Enter the number"))

  elif guess_numb < numb:
    print("YOUR GUESS IS LOW, TRY AGAIN")
    guess_numb = int(input("Enter the number"))
```
